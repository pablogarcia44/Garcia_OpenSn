```python
import h5py
import matplotlib.pyplot as plt

def explorer_et_tracer_fichier_h5(nom_fichier):
    try:
        # Ouvrir le fichier HDF5 en mode lecture
        with h5py.File(nom_fichier, 'r') as fichier:
            # Fonction pour explorer les groupes et datasets
            def explorer_groupe(groupe, chemin='/'):
                for cle in groupe.keys():
                    chemin_complet = chemin + cle
                    if isinstance(groupe[cle], h5py.Group):
                        print(f"Groupe: {chemin_complet}/")
                        explorer_groupe(groupe[cle], chemin_complet + '/')
                    elif isinstance(groupe[cle], h5py.Dataset):
                        print(f"Dataset: {chemin_complet}, Shape: {groupe[cle].shape}, Type: {groupe[cle].dtype}")
                        tracer_dataset(groupe[cle], chemin_complet)
            
            # Fonction pour tracer un dataset
            def tracer_dataset(dataset, nom_dataset):
                try:
                    data = dataset[:]  # Lire le dataset en tant que tableau numpy
                    plt.figure()
                    if data.ndim == 1:
                        plt.plot(data)
                        plt.xlabel('Indice groupe')
                        plt.ylabel('MGXS, cm-1')
                    elif data.ndim == 2:
                        plt.imshow(data, aspect='auto')
                        plt.colorbar()
                        plt.xlabel('Indice groupe')
                        plt.ylabel('MGXS, cm-1')
                    plt.title(f'Données du dataset {nom_dataset}')
                    plt.show()
                except Exception as e:
                    print(f"Erreur lors du tracé du dataset {nom_dataset}: {e}")
            
            # Explorer le fichier et tracer les datasets
            explorer_groupe(fichier)
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier: {e}")

# Nom du fichier .h5 à lire
nom_fichier = 'votre_fichier.h5'

# Appel de la fonction pour explorer et tracer les datasets
explorer_et_tracer_fichier_h5(nom_fichier)
```

```python
import h5py

def explorer_et_creer_dictionnaire_fichier_h5(nom_fichier):
    try:
        # Ouvrir le fichier HDF5 en mode lecture
        with h5py.File(nom_fichier, 'r') as fichier:
            valeurs_datasets = {}  # Dictionnaire pour stocker les valeurs des datasets
            
            # Fonction pour explorer les groupes et datasets
            def explorer_groupe(groupe, chemin='/'):
                for cle in groupe.keys():
                    chemin_complet = chemin + cle
                    if isinstance(groupe[cle], h5py.Group):
                        explorer_groupe(groupe[cle], chemin_complet + '/')
                    elif isinstance(groupe[cle], h5py.Dataset):
                        recuperer_valeurs_dataset(groupe[cle], chemin_complet)
            
            # Fonction pour récupérer les valeurs d'un dataset
            def recuperer_valeurs_dataset(dataset, nom_dataset):
                try:
                    data = dataset[:]  # Lire le dataset en tant que tableau numpy
                    valeurs_datasets[nom_dataset] = data.tolist()  # Convertir en liste et stocker
                except Exception as e:
                    print(f"Erreur lors de la récupération des valeurs du dataset {nom_dataset}: {e}")
            
            # Explorer le fichier et récupérer les valeurs des datasets
            explorer_groupe(fichier)
            
            return valeurs_datasets
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier: {e}")
        return None

# Nom du fichier .h5 à lire
nom_fichier = 'mgxs_1_10.h5'

# Appel de la fonction pour explorer et créer le dictionnaire
valeurs_donnees = explorer_et_creer_dictionnaire_fichier_h5(nom_fichier)

# Vérification et récupération des valeurs avec une clé spécifique
cle_recherchee = '/clad_guide/294K/scatter_data/g_min'
if valeurs_donnees is not None:
    liste_valeurs = valeurs_donnees.get(cle_recherchee)
    
    if liste_valeurs is not None:
        print(f"Valeurs pour la clé '{cle_recherchee}': {liste_valeurs}")
    else:
        print(f"La clé '{cle_recherchee}' n'existe pas dans le dictionnaire.")
else:
    print("Aucune donnée n'a été récupérée du fichier HDF5.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Supposons que vos listes soient nommées ainsi :
g_min_list = valeurs_donnees['/moderator_pincell/294K/scatter_data/g_min']  # Liste des gprime minimum pour chaque g
g_max_list = valeurs_donnees['/moderator_pincell/294K/scatter_data/g_max']  # Liste des gprime maximum pour chaque g
sigma_scat_list = valeurs_donnees['/moderator_pincell/294K/scatter_data/scatter_matrix']  # Liste des valeurs des sections efficaces de g vers gprime

# Nombre total de groupes
num_groups = len(g_min_list)

# Trouver la taille maximale pour gprime
max_gprime = max(g_max_list) + 1  # Assurez-vous que la taille est suffisante pour inclure tous les indices

# Initialisation de la matrice de diffusion
diffusion_matrix = np.zeros((max_gprime, num_groups))

# Remplissage de la matrice
for g in range(num_groups):
    min_gprime = g_min_list[g]
    max_gprime = g_max_list[g]
    # Assurez-vous que les indices sont dans les limites de la matrice
    if min_gprime < diffusion_matrix.shape[0] and max_gprime < diffusion_matrix.shape[0]:
        diffusion_matrix[min_gprime:max_gprime + 1, g] = sigma_scat_list[g]
    else:
        print(f"Indices hors limites pour g = {g}: min_gprime = {min_gprime}, max_gprime = {max_gprime}")

# Création d'une colormap personnalisée
def create_colormap():
    # Définir les couleurs : blanc pour 0, jaune foncé pour les faibles valeurs, jaune, orange, rouge
    colors = [(1, 1, 1), (1, 1, 0.2), (1, 1, 0), (1, 0.65, 0), (1, 0, 0)]  # Blanc, jaune foncé, jaune, orange, rouge
    n_bins = 100  # Nombre de segments dans la colormap
    cmap = mcolors.LinearSegmentedColormap.from_list('white_yellow_red_cmap', colors, N=n_bins)
    return cmap

# Créez la colormap personnalisée
cmap = create_colormap()

# Ajuster les seuils pour la normalisation logarithmique
# Pour éviter les problèmes avec les zéros, nous ajoutons une petite valeur epsilon
epsilon = 1e-10
norm = mcolors.LogNorm(vmin=epsilon, vmax=np.max(diffusion_matrix))

# Tracé de la matrice avec la colormap personnalisée
plt.figure(figsize=(10, 8))  # Ajuster la taille de la figure pour plus de clarté
plt.imshow(diffusion_matrix, cmap=cmap, norm=norm, origin='upper', interpolation='none')

# Ajouter une colorbar avec une taille de tick plus grande
colorbar = plt.colorbar(label='Section efficace de diffusion, cm-1')
colorbar.ax.tick_params(labelsize=12)  # Taille des labels de la colorbar

# Ajuster les labels des axes et la taille des ticks
plt.xlabel('g', fontsize=14)
plt.ylabel('gprime', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Titre avec une taille de police plus grande
plt.title('Matrice de diffusion neutronique dans le moderateur', fontsize=16)

plt.show()
```
