### Create function to plot

```python
import matplotlib.pyplot as plt

def plot_graph_with_custom_titles(keff1, X1, keff2=None, X2=None, keff3=None, X3=None, reference_value=1.18829, 
                                   label1='Dataset 1', label2='Dataset 2', label3='Dataset 3',
                                   xlabel='X (x-values)', ylabel='keff (y-values)', 
                                   title='Graph of keff vs X with Reference Line', 
                                   reference_label='Reference Value'):
    """
    Plots one, two, or three datasets on the same graph with a horizontal reference line and custom labels.
    
    :param keff1: List of y-values for the first dataset
    :param X1: List of x-values for the first dataset
    :param keff2: List of y-values for the second dataset (optional)
    :param X2: List of x-values for the second dataset (optional)
    :param keff3: List of y-values for the third dataset (optional)
    :param X3: List of x-values for the third dataset (optional)
    :param reference_value: The y-value where the reference line should be drawn
    :param label1: Label for the first dataset
    :param label2: Label for the second dataset
    :param label3: Label for the third dataset
    :param xlabel: Label for the x-axis
    :param ylabel: Label for the y-axis
    :param title: Title of the plot
    :param reference_label: Label for the reference line
    """
    # Check that the lists have the same length if they are provided
    if len(keff1) != len(X1):
        raise ValueError("The lists keff1 and X1 must have the same length.")
    
    if keff2 is not None and X2 is not None:
        if len(keff2) != len(X2):
            raise ValueError("The lists keff2 and X2 must have the same length.")
    
    if keff3 is not None and X3 is not None:
        if len(keff3) != len(X3):
            raise ValueError("The lists keff3 and X3 must have the same length.")
    
    # Create the plot
    plt.figure(figsize=(10, 6))

    # Plot the first dataset
    plt.plot(X1, keff1, marker='o', linestyle='-', color='b', label=label1)

    # Plot the second dataset if provided
    if keff2 is not None and X2 is not None:
        plt.plot(X2, keff2, marker='s', linestyle='--', color='g', label=label2)

    # Plot the third dataset if provided
    if keff3 is not None and X3 is not None:
        plt.plot(X3, keff3, marker='^', linestyle='-.', color='m', label=label3)

    # Add the reference line
    plt.axhline(y=reference_value, color='r', linestyle='--', label=f'{reference_label} ')

    # Add titles and labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # Show grid and legend
    plt.grid(True)
    plt.legend()

    # Display the plot
    plt.show()

# Example usage of the function

# Single dataset with custom labels and titles
keff1 = [10, 20, 15, 30, 25, 35, 40]
X1 = [1, 2, 3, 4, 5, 6, 7]
plot_graph_with_custom_titles(
    keff1, X1, 
    xlabel='Time (s)', 
    ylabel='Temperature (°C)', 
    title='Temperature vs Time with Reference Line',
    reference_value=25,
    reference_label='Critical Temperature'
)

# Two datasets with custom labels and titles
keff2 = [5, 15, 10, 20, 18, 25, 30]
X2 = [1, 2, 3, 4, 5, 6, 7]
plot_graph_with_custom_titles(
    keff1, X1, keff2, X2, 
    label1='Experiment A', 
    label2='Experiment B', 
    xlabel='Measurement Point', 
    ylabel='Value', 
    title='Comparison of Two Experiments with Reference Line',
    reference_value=18,
    reference_label='Threshold'
)

# Three datasets with custom labels and titles
keff3 = [3, 8, 12, 15, 22, 27, 35]
X3 = [1, 2, 3, 4, 5, 6, 7]
plot_graph_with_custom_titles(
    keff1, X1, keff2, X2, keff3, X3, 
    label1='Experiment A', 
    label2='Experiment B',
    label3='Experiment C',
    xlabel='Measurement Point', 
    ylabel='Value', 
    title='Comparison of Three Experiments with Reference Line',
    reference_value=18,
    reference_label='Threshold'
)
```

### Create function to plot + point name 

```python
import matplotlib.pyplot as plt

def plot_graph_with_custom_titles2(keff1, X1, labels1=None, keff2=None, X2=None, labels2=None, keff3=None, X3=None, labels3=None,
                                   reference_value=1.18829, 
                                   label1='Dataset 1', label2='Dataset 2', label3='Dataset 3',
                                   xlabel='X (x-values)', ylabel='keff (y-values)', 
                                   title='Graph of keff vs X with Reference Line', 
                                   reference_label='Reference Value'):
    """
    Plots one, two, or three datasets on the same graph with a horizontal reference line and custom labels.
    Optionally displays labels next to each point.
    
    :param keff1: List of y-values for the first dataset
    :param X1: List of x-values for the first dataset
    :param labels1: List of strings for labels to be displayed next to points in the first dataset (optional)
    :param keff2: List of y-values for the second dataset (optional)
    :param X2: List of x-values for the second dataset (optional)
    :param labels2: List of strings for labels to be displayed next to points in the second dataset (optional)
    :param keff3: List of y-values for the third dataset (optional)
    :param X3: List of x-values for the third dataset (optional)
    :param labels3: List of strings for labels to be displayed next to points in the third dataset (optional)
    :param reference_value: The y-value where the reference line should be drawn
    :param label1: Label for the first dataset
    :param label2: Label for the second dataset
    :param label3: Label for the third dataset
    :param xlabel: Label for the x-axis
    :param ylabel: Label for the y-axis
    :param title: Title of the plot
    :param reference_label: Label for the reference line
    """
    # Check that the lists have the same length if they are provided
    if len(keff1) != len(X1):
        raise ValueError("The lists keff1 and X1 must have the same length.")
    if labels1 is not None and len(labels1) != len(X1):
        raise ValueError("The list labels1 must have the same length as keff1 and X1.")
    
    if keff2 is not None and X2 is not None:
        if len(keff2) != len(X2):
            raise ValueError("The lists keff2 and X2 must have the same length.")
        if labels2 is not None and len(labels2) != len(X2):
            raise ValueError("The list labels2 must have the same length as keff2 and X2.")
    
    if keff3 is not None and X3 is not None:
        if len(keff3) != len(X3):
            raise ValueError("The lists keff3 and X3 must have the same length.")
        if labels3 is not None and len(labels3) != len(X3):
            raise ValueError("The list labels3 must have the same length as keff3 and X3.")
    
    # Create the plot
    plt.figure(figsize=(10, 6))

    # Plot the first dataset
    plt.plot(X1, keff1, marker='o', linestyle='-', color='b', label=label1)
    if labels1 is not None:
        for i in range(len(X1)):
            plt.text(X1[i], keff1[i], labels1[i], fontsize=9, ha='right')

    # Plot the second dataset if provided
    if keff2 is not None and X2 is not None:
        plt.plot(X2, keff2, marker='s', linestyle='--', color='g', label=label2)
        if labels2 is not None:
            for i in range(len(X2)):
                plt.text(X2[i], keff2[i], labels2[i], fontsize=9, ha='right')

    # Plot the third dataset if provided
    if keff3 is not None and X3 is not None:
        plt.plot(X3, keff3, marker='^', linestyle='-.', color='m', label=label3)
        if labels3 is not None:
            for i in range(len(X3)):
                plt.text(X3[i], keff3[i], labels3[i], fontsize=9, ha='right')

    # Add the reference line
    plt.axhline(y=reference_value, color='r', linestyle='--', label=f'{reference_label} ')

    # Add titles and labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # Show grid and legend
    plt.grid(True)
    plt.legend()

    # Display the plot
    plt.show()

# Example usage of the function with labels for points
keff1 = [10, 20, 15, 30, 25, 35, 40]
X1 = [1, 2, 3, 4, 5, 6, 7]
labels1 = ["Point A", "Point B", "Point C", "Point D", "Point E", "Point F", "Point G"]
plot_graph_with_custom_titles2(
    keff1, X1, labels1, 
    xlabel='Time (s)', 
    ylabel='Temperature (°C)', 
    title='Temperature vs Time with Labels',
    reference_value=25,
    reference_label='Critical Temperature'
)
```
